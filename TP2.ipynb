```python
from sklearn import *
import numpy as np
from matplotlib.pyplot import *

X = np.array([[1., -1., 2.], [2., 0., 0.], [0., 1., -1.]])
print(" Matrice X")
print(X)

print("\n Moyenne de X")
print(X.mean(0))

print("\n Variance de X")
print(X.std(0))

print("\n Matrice X normalisée")
X_scaled = preprocessing.scale(X)
print(X_scaled)

print("\n Moyenne de X normalisée")
print(X_scaled.mean(0))

print("\n Variance de X normalisée")
print(X_scaled.std(0))

print("\n --------------------------------------------------------")

X2 = np.array([[1., -1., 2.], [2., 0., 0.], [0., 1., -1.]])
print("\n Matrice X2")
print(X2)

X2_minMax = preprocessing.MinMaxScaler().fit_transform(X2)
print(" \n Matrice X2_minMax")
print(X2_minMax)

print("\n Moyenne de X2_minMax")
print(X2_minMax.mean(0))

print("\n Variance de X2_minMax")
print(X2_minMax.std(0))

print("\n --------------------------------------------------------")

iris = datasets.load_iris()

figure()
scatter(iris.data[:, 0], iris.data[:, 1], c=iris.target, s=10)
title("Affichage des donnees data = 0 : 1")
xlabel("Axe des X")
ylabel("Axe des Y")
show()

figure()
scatter(iris.data[:, 0], iris.data[:, 2], c=iris.target, s=10)
title("Affichage des donnees data = 0 : 2")
xlabel("Axe des X")
ylabel("Axe des Y")
show()

figure()
scatter(iris.data[:, 0], iris.data[:, 3], c=iris.target, s=10)
title("Affichage des donnees data = 0 : 3")
xlabel("Axe des X")
ylabel("Axe des Y")
show()

figure()
scatter(iris.data[:, 1], iris.data[:, 2], c=iris.target, s=10)
title("Affichage des donnees data = 1 : 2")
xlabel("Axe des X")
ylabel("Axe des Y")
show()

figure()
scatter(iris.data[:, 1], iris.data[:, 3], c=iris.target, s=10)
title("Affichage des donnees data = 1 : 3")
xlabel("Axe des X")
ylabel("Axe des Y")
show()

figure()
scatter(iris.data[:, 2], iris.data[:, 3], c=iris.target, s=10)
title("Affichage des donnees data = 2 : 3")
xlabel("Axe des X")
ylabel("Axe des Y")
show()

print("\n Matrice avec la fonction corrcoef")
print(np.corrcoef(iris.data))

print("\n --------------------------------------------------------")

from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

iris = datasets.load_iris()

X = iris.data
y = iris.target

target_names = iris.target_names
pca = PCA(n_components=2)
IrisPCA = pca.fit(X,y).transform(X)
lda = LinearDiscriminantAnalysis(n_components=2)
IrisLDA = lda.fit(X,y).transform(X)

figure()
scatter(IrisPCA.data[:, 0], IrisPCA.data[:, 1], s=10)
title("Affichage des donnees PCA")
xlabel("Axe des X")
ylabel("Axe des Y")
show()
```
